Visualization

```python
import matplotlib.pyplot as plt
```

```python
## Calculating Total Amount
Total_amount=df['transaction_qty']*df['unit_price']
df['Total_amount']=Total_amount
df.head(2)
```

```python
## Revenue by product
revenue_by_product=df.groupby('product_category')['Total_amount'].sum()
print(revenue_by_product)
revenue_by_product.plot(kind='bar')
plt.xlabel('Product ID')
plt.ylabel('Revenue')
plt.title('Revenue by Product')
plt.show()

```

```python
df['transaction_time'] = pd.to_datetime(df['transaction_time'])
df["hour"]=df["transaction_time"].dt.hour
df["day"]=df["transaction_time"].dt.day
df["month"]=df["transaction_time"].dt.month
df["year"]=df["transaction_time"].dt.year
df.head(2)
```

```python
revenue_over_time=df.groupby('revenue_by_product')['	transaction_date'].sum()
revenue_over_time.plot(kind='line')
plt.xlabel('Revenue_Over_Time')
plt.ylabel('Date')
plt.title('Revenue')
plt.show()
```
